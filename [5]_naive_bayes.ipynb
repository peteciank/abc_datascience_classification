# Naive Bayes

```
# Naive Bayes

## Importing the libraries

import numpy as np  # For numerical computations
import matplotlib.pyplot as plt  # For data visualization
import pandas as pd  # For dataset manipulation

## Importing the dataset

dataset = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-GzTqfan25hVJ4BrsZr0dimNorFVk7OQ49yGl39SWEEtL8pZ6qlSzZczV2jhg32Zni8AiPw0B3e-W/pub?gid=1606731410&single=true&output=csv')
X = dataset.iloc[:, :-1].values  # Extract the independent variables (features)
y = dataset.iloc[:, -1].values  # Extract the dependent variable (target class)

## Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
# Splits the data into 75% training and 25% testing sets

## Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()  # Create a standard scaler object
X_train = sc.fit_transform(X_train)  # Scale the training features
X_test = sc.transform(X_test)  # Scale the testing features

## Training the Naive Bayes model on the Training set

from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()  # Create a Gaussian Naive Bayes classifier
classifier.fit(X_train, y_train)  # Train the model on the training data

## Making the Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)  # Generate predictions on the test set
cm = confusion_matrix(y_test, y_pred)  # Create the confusion matrix
print(cm)
accuracy_score(y_test, y_pred)  # Calculate the accuracy of the model
```

**Key Points**:

* **Independence Assumption**: Naive Bayes assumes that all features are independent of each other, given the class variable. This assumption is often unrealistic but can still lead to good results in certain cases.

* **GaussianNB**: This code uses the `GaussianNB` classifier, which is suitable when your features follow a Gaussian (normal) distribution.

* **Other Naive Bayes Variants**: Explore other variants like `MultinomialNB` (for discrete features) and `BernoulliNB` (for binary features).
