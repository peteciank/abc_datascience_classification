# Logistic Regression

```
# Logistic Regression

## Importing the libraries

import numpy as np  # For array operations and mathematical functions
import matplotlib.pyplot as plt  # For data visualization
import pandas as pd  # For data handling and analysis

## Importing the dataset

dataset = pd.read_csv('https://docs.google.com/spreadsheets/d/e/2PACX-1vQ-GzTqfan25hVJ4BrsZr0dimNorFVk7OQ49yGl39SWEEtL8pZ6qlSzZczV2jhg32Zni8AiPw0B3e-W/pub?gid=1606731410&single=true&output=csv')
X = dataset.iloc[:, :-1].values  # Extract feature variables
y = dataset.iloc[:, -1].values  # Extract the target variable (class labels)

## Splitting the dataset into the Training set and Test set

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=0)
# Splits data into 75% training/ 25% testing with random_state for consistent results

## Feature Scaling

from sklearn.preprocessing import StandardScaler
sc = StandardScaler() # Create a scaler object
X_train = sc.fit_transform(X_train)  # Standardize the training set
X_test = sc.transform(X_test)  # Standardize the testing set using the same scaler

## Training the Logistic Regression model on the Training set

from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression(random_state=0)
classifier.fit(X_train, y_train)  # Train the model on the training data

## Making the Confusion Matrix

from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = classifier.predict(X_test)  # Generate predictions on the test set
cm = confusion_matrix(y_test, y_pred)  # Create the confusion matrix
print(cm)
accuracy_score(y_test, y_pred)  # Calculate the model's accuracy
```

**Key Points**:

* **Assumptions**: Logistic regression assumes a linear relationship between the input features and the log-odds of the target variable.

* **Regularization**: Explore regularization parameters (penalty and C) in the LogisticRegression function to potentially prevent overfitting.
